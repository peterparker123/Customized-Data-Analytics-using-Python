```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(t):
    'A damped exponential'
    s1 = np.cos(2 * np.pi * t)
    e1 = np.exp(-t)
    return s1 * e1
```

```python
t1 = np.arange(0.0, 5.0, .2)
```

```python
l = plt.plot(t1, f(t1), 'ro')
plt.setp(l)
plt.setp(l, markersize=30)
plt.setp(l, markerfacecolor='C2')
```
