## The `chop_threshold()` Option

```python

pd.reset_option("precision")
pd.options.display.precision
df = pd.DataFrame(np.random.randn(10, 10))
df
```

```python

pd.set_option("display.chop_threshold", 1)
df
```

```python
pd.reset_option("chop_threshold")
```
